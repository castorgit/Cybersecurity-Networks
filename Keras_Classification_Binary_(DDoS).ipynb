**Keras - Binary Classification on DDoS2007 file**

```python
%matplotlib notebook 
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras import Sequential
from keras.layers import Dense
from keras import backend as K
import tensorflow as tf

tf.logging.set_verbosity(tf.logging.ERROR)
```

    Some functions to transform data or plot confusion matrix

```python
def spounge(string):
   new = ['0' for i in range(43)]
   a = string
   l = a.split()
   for i in l:
     n = int(i.split(":")[0])
     if len(i) == 1:
        new[0] = str(n)
     else:
        new[n] = i.split(':')[1]
   return new

def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=True,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

    read file into dataframe

```python
df = pd.read_csv('e:/dalhousie_spark/DDoS2007.csv', header = None)
```

    transformation dataframe into columns (each packet one column)

```python
tmp = df.apply(lambda row: (spounge(row[0])),axis=1)
m = pd.DataFrame(tmp.tolist())
m = m.astype("int64")
```

    Dataset regularization and Split

```python
# creating input columns and target column
X= m.iloc[:,1:42]
y= m.iloc[:,0]

 #regularizing the dataset (works better)
 #standardizing  input features

sc = StandardScaler()
X = sc.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

    The Network Model (42:128:128:1)

```python
NN = Sequential()
NN.add(Dense(128, activation='relu', kernel_initializer='random_normal', input_dim=41))
NN.add(Dense(128, activation='relu', kernel_initializer='random_normal'))
NN.add(Dense(1, activation='sigmoid', kernel_initializer='random_normal'))
```

    Train and Fit

```python
NN.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
NN.summary()
#Fitting the data to the training dataset
NN.fit(X_train,y_train, batch_size=10, epochs=1)

eval_model=NN.evaluate(X_train, y_train)
print("model result: ",eval_model)
```

    Final evaluation / Confusion Matrix

```python
y_pred = NN.predict(X_test)
y_pred =(y_pred>0.5)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

    Confusion Plots

```python
plot_confusion_matrix(y_test, y_pred, classes=np.array(['0','1']),
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues)
plot_confusion_matrix(y_test, y_pred, classes=np.array(['0','1']),
                          normalize=True,
                          title=None,
                          cmap=plt.cm.Blues)
```
